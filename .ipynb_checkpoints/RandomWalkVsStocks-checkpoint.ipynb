```python
def listMean(arr,size):
    avg = []
    for i in range(size):
        acc = 0
        for j in range(len(arr)):
            acc += arr[j][i]
        avg.append(acc/len(arr))
    return avg
```

```python
def plotRandomWalk(lenght,init_val):
    walk = [init_val]
    walks = []
    for j in range(100):
        for i in range(lenght):
            chng = random.normalvariate(0.2269390716093324,4.824894466012708)
            walk.append(chng*abs(walk[-1])/100 + walk[-1])
        walks.append(walk)
    plt.plot(listMean(walks,lenght))
```
